```python
import xml.etree.ElementTree as ET
import pandas as pd

# === 1️⃣ Wczytanie pliku XML ===
file_path = "sewik.xml"
tree = ET.parse(file_path)
root = tree.getroot()

# === 2️⃣ Przygotowanie list na dane ===
zdarzenia_list = []
pojazdy_list = []
uczestnicy_list = []

# === 3️⃣ Iteracja po wszystkich zdarzeniach ===
for zdarzenie in root.find("ZDARZENIA").findall("ZDARZENIE"):
    zd_id = zdarzenie.findtext("ID")
    data_zdarz = zdarzenie.findtext("DATA_ZDARZENIA")
    miejscowosc = zdarzenie.findtext("MIEJSCOWOSC")
    powiat = zdarzenie.findtext("POWIAT")
    gmina = zdarzenie.findtext("GMINA")
    woj = zdarzenie.findtext("WOJ")
    predkosc = zdarzenie.findtext("PREDKOSC_DOPUSZCZALNA")
    droga_publiczna = zdarzenie.findtext("DROGA_PUBLICZNA")

    # === Dodajemy zdarzenie do listy ===
    zdarzenia_list.append({
        "ID_ZDARZENIA": zd_id,
        "DATA_ZDARZENIA": data_zdarz,
        "WOJEWODZTWO": woj,
        "POWIAT": powiat,
        "GMINA": gmina,
        "MIEJSCOWOSC": miejscowosc,
        "PREDKOSC_DOPUSZCZALNA": predkosc,
        "DROGA_PUBLICZNA": droga_publiczna
    })

    # === Pojazdy ===
    pojazdy = zdarzenie.find("POJAZDY")
    if pojazdy is not None:
        for p in pojazdy.findall("POJAZD"):
            pojazdy_list.append({
                "ID_POJAZDU": p.findtext("ID"),
                "ID_ZDARZENIA": zd_id,
                "NR_POJAZDU": p.findtext("NR_POJAZDU"),
                "MARKA": p.findtext("MARKA"),
                "ROK_PRODUKCJI": p.findtext("ROK_PRODUKCJI"),
                "UBEZPIECZYCIEL": p.findtext("SPSU_KOD"),
                "DATA_OST_BAD_TECH": p.findtext("DATA_OST_BAD_TECH")
            })

    # === Uczestnicy ===
    uczestnicy = zdarzenie.find("UCZESTNICY")
    if uczestnicy is not None:
        for o in uczestnicy.findall("OSOBA"):
            uczestnicy_list.append({
                "ID_OSOBY": o.findtext("ID"),
                "ID_ZDARZENIA": zd_id,
                "ID_POJAZDU": o.findtext("ZSPO_ID"),
                "PLEC": o.findtext("PLEC"),
                "DATA_UR": o.findtext("DATA_UR"),
                "OBYWATELSTWO": o.findtext("SOBY_KOD"),
                "OBCOKRAJOWIEC": o.findtext("OBCOKRAJOWIEC"),
                "STATUS": o.findtext("SUSU_KOD"),
                "LAT_KIEROWANIA": o.findtext("LICZBA_LAT_KIEROWANIA"),
            })

# === 4️⃣ Konwersja do DataFrame ===
df_zdarzenia = pd.DataFrame(zdarzenia_list)
df_pojazdy = pd.DataFrame(pojazdy_list)
df_uczestnicy = pd.DataFrame(uczestnicy_list)

print(df_pojazdy)
```
